8. Метод is_point_inside (x, y, z), который принимает 3 аргумента — действительных числа — координаты некоторой точки в пространстве (в том же порядке, что и в конструкторе), и возвращает логическое значение True или False в зависимости от того, находится эта точка внутри сферы.

````python
import math

class Sphere:
    def __init__(self, *args):
        if len(args) == 0:
            self.radius = 1.0
            self.center = (0.0, 0.0, 0.0)
        elif len(args) == 1:
            self.radius = float(args[0])
            self.center = (0.0, 0.0, 0.0)
        elif len(args) == 4:
            self.radius = float(args[0])
            self.center = (float(args[1]), float(args[2]), float(args[3]))
        else:
            raise ValueError("Неправильное количество аргументов. Используйте 0, 1 или 4 аргумента.")

    def get_volume(self):
        return (4/3) * math.pi * (self.radius ** 3)

    def get_square(self):
        return 4 * math.pi * (self.radius ** 2)

    def get_radius(self):
        return self.radius

    def get_center(self):
        return self.center

    def set_radius(self, r):
        self.radius = float(r)

    def set_center(self, x, y, z):
        self.center = (float(x), float(y), float(z))

    def is_point_inside(self, x, y, z):
        dx = x - self.center[0]
        dy = y - self.center[1]
        dz = z - self.center[2]
        distance_squared = dx**2 + dy**2 + dz**2

        return distance_squared < self.radius**2


s0 = Sphere(0.5) # test sphere creation with radius and default center
print(s0.get_center()) # (0.0, 0.0, 0.0)
print(round(s0.get_volume(), 12)) # 0.523598775598
print(s0.is_point_inside(0 , -1.5, 0)) # False
s0.set_radius(1.6)
print(s0.is_point_inside(0, -1.5, 0)) # True
print(s0.get_radius()) # 1.6
````

## **Задание №2. Создайте класс SuperStr, который наследует стандартный класс str и содержит 2 новых метода:**

1. Метод is_repeatance (s), который принимает 1 аргумент s и возвращает True или False в зависимости от того, может ли текущая строку быть получена целым количеством повторов строки s. Вернуть False, если s не является строкой. Считать, что пустая строка не содержит повторов.

2. Метод is_palindrom (), который возвращает True или False в зависимости от того, является ли строка [палиндромом](https://ru.wikipedia.org/wiki/Палиндром). Регистрами символов пренебрегать. Пустую строку считать палиндромом.

````python
class SuperStr(str):
    def is_repeatance(self, s):
        if not isinstance(s, str):
            return False
        if not s:
            return False
        if len(self) % len(s) != 0:
            return False

        repeats = len(self) // len(s)
        return self == s * repeats

    def is_palindrom(self):
        text = self.lower()
        if not text:
            return True
        return text == text[::-1]


# Тесты
if __name__ == "__main__":
    s = SuperStr("123123123123")
    print(s.is_repeatance("123"))  # True
    print(s.is_repeatance("123123"))  # True
    print(s.is_repeatance("123123123123"))  # True
    print(s.is_repeatance("12312"))  # False
    print(s.is_repeatance(123))  # False
    print(s.is_palindrom())  # False
    print(s)  # 123123123123 (строка)
    print(int(s))  # 123123123123 (целое число)
    print(s + "qwe")  # 123123123123qwe
    p = SuperStr("123_321")
    print(p.is_palindrom())  # True
````

````python
class Student:
    def __init__(self, name, surname, gender):
        self.name = name
        self.surname = surname
        self.gender = gender
        self.finished_courses = []
        self.courses_in_progress = []
        self.grades = {}

    def __eq__(self, other):
        """Сравнение студентов по средней оценке"""
        if not isinstance(other, Student):
            return NotImplemented
        return self.average_grade() == other.average_grade()

    def __lt__(self, other):
        """Сравнение студентов: кто имеет меньшую среднюю оценку"""
        if not isinstance(other, Student):
            return NotImplemented
        return self.average_grade() < other.average_grade()

    def __le__(self, other):
        """Сравнение студентов: меньше или равно по средней оценке"""
        if not isinstance(other, Student):
            return NotImplemented
        return self.average_grade() <= other.average_grade()

    def __len__(self):
        """Количество оценок у студента"""
        total_grades = 0
        for grades_list in self.grades.values():
            total_grades += len(grades_list)
        return total_grades

    def average_grade(self):
        """Вычисление средней оценки (вспомогательный метод)"""
        all_grades = []
        for course_grades in self.grades.values():
            all_grades.extend(course_grades)
        return sum(all_grades) / len(all_grades) if all_grades else 0

    def __str__(self):
        courses_in_progress_str = ', '.join(self.courses_in_progress) if self.courses_in_progress else 'Нет курсов'
        finished_courses_str = ', '.join(self.finished_courses) if self.finished_courses else 'Нет курсов'

        return (f"Имя: {self.name}\n"
                f"Фамилия: {self.surname}\n"
                f"Средняя оценка за домашние задания: {self.average_grade():.1f}\n"
                f"Курсы в процессе изучения: {courses_in_progress_str}\n"
                f"Завершенные курсы: {finished_courses_str}")

    def add_courses(self, course_name):
        self.finished_courses.append(course_name)

    def rate_lecturer(self, lecturer, course, grade):
        if not isinstance(lecturer, Lecturer):
            return 'Ошибка: это не лектор'

        if course not in lecturer.courses_attached:
            return 'Ошибка: лектор не преподает на этом курсе'

        if course not in self.courses_in_progress and course not in self.finished_courses:
            return 'Ошибка: студент не записан на этот курс'

        if not 1 <= grade <= 10:
            return 'Ошибка: оценка должна быть от 1 до 10'

        if course in lecturer.lecturer_grades:
            lecturer.lecturer_grades[course].append(grade)
        else:
            lecturer.lecturer_grades[course] = [grade]

        return 'Оценка успешно добавлена'


class Mentor:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
        self.courses_attached = []

    def __str__(self):
        return f"Имя: {self.name}\nФамилия: {self.surname}"

    def __eq__(self, other):
        """Сравнение менторов по именам"""
        if not isinstance(other, Mentor):
            return NotImplemented
        return (self.name, self.surname) == (other.name, other.surname)


class Lecturer(Mentor):
    def __init__(self, name, surname):
        super().__init__(name, surname)
        self.lecturer_grades = {}

    def __eq__(self, other):
        """Сравнение лекторов по средней оценке"""
        if not isinstance(other, Lecturer):
            return NotImplemented
        return self.average_grade() == other.average_grade()

    def __lt__(self, other):
        """Сравнение лекторов: кто имеет меньшую среднюю оценку"""
        if not isinstance(other, Lecturer):
            return NotImplemented
        return self.average_grade() < other.average_grade()

    def __len__(self):
        """Количество лекций (курсов), которые ведет лектор"""
        return len(self.courses_attached)

    def average_grade(self):
        """Вычисление средней оценки лектора"""
        all_grades = []
        for course_grades in self.lecturer_grades.values():
            all_grades.extend(course_grades)
        return sum(all_grades) / len(all_grades) if all_grades else 0

    def __str__(self):
        return (f"Имя: {self.name}\n"
                f"Фамилия: {self.surname}\n"
                f"Средняя оценка за лекции: {self.average_grade():.1f}\n"
                f"Количество читаемых курсов: {len(self)}")


class Reviewer(Mentor):
    def __init__(self, name, surname):
        super().__init__(name, surname)

    def __len__(self):
        """Количество курсов, которые проверяет ревьюер"""
        return len(self.courses_attached)

    def __contains__(self, course):
        """Проверка, проверяет ли ревьюер определенный курс"""
        return course in self.courses_attached

    def rate_hw(self, student, course, grade):
        if isinstance(student, Student) and course in self.courses_attached and course in student.courses_in_progress:
            if course in student.grades:
                student.grades[course] += [grade]
            else:
                student.grades[course] = [grade]
            return 'Оценка успешно добавлена'
        else:
            return 'Ошибка'


if __name__ == "__main__":
    best_student = Student('Ruoy', 'Eman', 'male')
    good_student = Student('Alice', 'Smith', 'female')
    average_student = Student('Bob', 'Johnson', 'male')

    cool_reviewer = Reviewer('Some', 'Buddy')
    strict_reviewer = Reviewer('Strict', 'Teacher')
    cool_lecturer = Lecturer('John', 'Doe')
    experienced_lecturer = Lecturer('Anna', 'Professor')

    best_student.courses_in_progress.extend(['Python', 'Git'])
    good_student.courses_in_progress.extend(['Python', 'Git', 'Algorithms'])
    average_student.courses_in_progress.extend(['Python'])

    best_student.add_courses('Введение в программирование')
    good_student.add_courses('Математика для программистов')

    cool_reviewer.courses_attached.extend(['Python', 'Git'])
    strict_reviewer.courses_attached.extend(['Algorithms'])

    cool_lecturer.courses_attached.extend(['Python', 'Git'])
    experienced_lecturer.courses_attached.extend(['Algorithms', 'Python'])

    print("=== Выставление оценок ===")
    print(cool_reviewer.rate_hw(best_student, 'Python', 9))
    print(cool_reviewer.rate_hw(best_student, 'Python', 10))
    print(cool_reviewer.rate_hw(best_student, 'Git', 8))

    print(cool_reviewer.rate_hw(good_student, 'Python', 7))
    print(cool_reviewer.rate_hw(good_student, 'Git', 8))
    print(strict_reviewer.rate_hw(good_student, 'Algorithms', 9))

    print(cool_reviewer.rate_hw(average_student, 'Python', 6))
    print(cool_reviewer.rate_hw(average_student, 'Python', 7))

    print(best_student.rate_lecturer(cool_lecturer, 'Python', 9))
    print(good_student.rate_lecturer(cool_lecturer, 'Python', 8))
    print(best_student.rate_lecturer(cool_lecturer, 'Git', 10))

    print(good_student.rate_lecturer(experienced_lecturer, 'Algorithms', 9))
    print(good_student.rate_lecturer(experienced_lecturer, 'Python', 8))

    print("\n" + "="*50)
    print("=== ДЕМОНСТРАЦИЯ МАГИЧЕСКИХ МЕТОДОВ ===")

    print("\n1. Сравнение студентов по средней оценке:")
    print(f"best_student == good_student: {best_student == good_student}")
    print(f"best_student < good_student: {best_student < good_student}")
    print(f"best_student <= good_student: {best_student <= good_student}")
    print(f"average_student < best_student: {average_student < best_student}")

    print(f"\n2. Количество оценок у студентов:")
    print(f"best_student: {len(best_student)} оценок")
    print(f"good_student: {len(good_student)} оценок")
    print(f"average_student: {len(average_student)} оценок")

    print(f"\n3. Количество курсов у лекторов:")
    print(f"cool_lecturer читает {len(cool_lecturer)} курса")
    print(f"experienced_lecturer читает {len(experienced_lecturer)} курсов")

    print(f"\n4. Проверка курсов у ревьюеров:")
    print(f"'Python' в курсах cool_reviewer: {'Python' in cool_reviewer}")
    print(f"'Algorithms' в курсах cool_reviewer: {'Algorithms' in cool_reviewer}")
    print(f"'Git' в курсах strict_reviewer: {'Git' in strict_reviewer}")

    print(f"\n5. Сравнение лекторов по средней оценке:")
    print(f"cool_lecturer == experienced_lecturer: {cool_lecturer == experienced_lecturer}")
    print(f"cool_lecturer < experienced_lecturer: {cool_lecturer < experienced_lecturer}")

    print(f"\n6. Сортировка студентов по средней оценке (от лучшего к худшему):")
    students = [best_student, good_student, average_student]
    sorted_students = sorted(students, reverse=True)

    for i, student in enumerate(sorted_students, 1):
        print(f"{i}. {student.name} {student.surname}: {student.average_grade():.1f}")

    print("\n" + "="*50)
    print("=== ИНФОРМАЦИЯ О ВСЕХ УЧАСТНИКАХ ===")

    print("\nСтуденты:")
    print("---------")
    for student in students:
        print(student)
        print()

    print("\nЛекторы:")
    print("--------")
    lecturers = [cool_lecturer, experienced_lecturer]
    for lecturer in lecturers:
        print(lecturer)
        print()

    print("\nРевьюеры:")
    print("---------")
    reviewers = [cool_reviewer, strict_reviewer]
    for reviewer in reviewers:
        print(reviewer)
        print(f"Количество проверяемых курсов: {len(reviewer)}")
        print()
````

   - **Вычитание `-`** (`__sub__`): вычитание одной дроби из другой, возвращает новую дробь.

   - **Официальное представление** (`__repr__`): возвращает строку вида `"Fraction(числитель, знаменатель)"`.

5. **Обработка исключений:**

   - При попытке создать дробь с нулевым знаменателем выводить в консоль сообщение `ValueError`.

   - Обеспечить корректную обработку исключений и вывод соответствующих сообщений.





````python
import math


class Fraction:
    def __init__(self, numerator, denominator=1):
        if denominator == 0:
            raise ValueError('Знаменатель не может быть нулем')
        self._simplify(numerator, denominator)

    def _simplify(self, numerator, denominator):
        if denominator < 0:
            numerator = -numerator
            denominator = -denominator
        gcd_value = math.gcd(abs(numerator), denominator)
        self.numerator = numerator // gcd_value
        self.denominator = denominator // gcd_value

    def __add__(self, other):
        if isinstance(other, Fraction):
            new_numerator = (self.numerator * other.denominator + other.numerator * self.denominator)
            new_denominator = self.denominator * other.denominator
            return Fraction(new_numerator, new_denominator)
        elif isinstance(other, (int, float)):
            return self + Fraction.from_float(other)
        else:
            return NotImplemented

    def __radd__(self, other):
        return self + other

    def __sub__(self, other):
        if isinstance(other, Fraction):
            new_numerator = (self.numerator * other.denominator -
                           other.numerator * self.denominator)
            new_denominator = self.denominator * other.denominator
            return Fraction(new_numerator, new_denominator)
        elif isinstance(other, (int, float)):
            return self - Fraction.from_float(other)
        else:
            return NotImplemented

    def __rsub__(self, other):
        if isinstance(other, (int, float)):
            return Fraction.from_float(other) - self
        else:
            return NotImplemented

    def __mul__(self, other):
        if isinstance(other, Fraction):
            new_numerator = self.numerator * other.numerator
            new_denominator = self.denominator * other.denominator
            return Fraction(new_numerator, new_denominator)
        elif isinstance(other, (int, float)):
            return self * Fraction.from_float(other)
        else:
            return NotImplemented

    def __rmul__(self, other):
        return self * other

    def __truediv__(self, other):
        if isinstance(other, Fraction):
            if other.numerator == 0:
                raise ZeroDivisionError('Деление на нулевую дробь')
            new_numerator = self.numerator * other.denominator
            new_denominator = self.denominator * other.numerator
            return Fraction(new_numerator, new_denominator)
        elif isinstance(other, (int, float)):
            if other == 0:
                raise ZeroDivisionError('Деление на ноль')
            return self / Fraction.from_float(other)
        else:
            return NotImplemented

    def __rtruediv__(self, other):
        if isinstance(other, (int, float)):
            return Fraction.from_float(other) / self
        else:
            return NotImplemented

    def __eq__(self, other):
        if isinstance(other, Fraction):
            return (self.numerator == other.numerator and
                   self.denominator == other.denominator)
        elif isinstance(other, (int, float)):
            return float(self) == other
        else:
            return NotImplemented

    def __lt__(self, other):
        if isinstance(other, Fraction):
            return (self.numerator * other.denominator <
                   other.numerator * self.denominator)
        elif isinstance(other, (int, float)):
            return float(self) < other
        else:
            return NotImplemented

    def __le__(self, other):
        if isinstance(other, Fraction):
            return (self.numerator * other.denominator <=
                   other.numerator * self.denominator)
        elif isinstance(other, (int, float)):
            return float(self) <= other
        else:
            return NotImplemented

    def __gt__(self, other):
        if isinstance(other, Fraction):
            return (self.numerator * other.denominator >
                   other.numerator * self.denominator)
        elif isinstance(other, (int, float)):
            return float(self) > other
        else:
            return NotImplemented

    def __ge__(self, other):
        if isinstance(other, Fraction):
            return (self.numerator * other.denominator >=
                   other.numerator * self.denominator)
        elif isinstance(other, (int, float)):
            return float(self) >= other
        else:
            return NotImplemented

    def __neg__(self):
        return Fraction(-self.numerator, self.denominator)

    def __float__(self):
        return self.numerator / self.denominator

    def reciprocal(self):
        if self.numerator == 0:
            raise ZeroDivisionError('Нельзя получить обратную дробь для нуля')
        return Fraction(self.denominator, self.numerator)

    def __str__(self):
        if self.denominator == 1:
            return str(self.numerator)
        return f"{self.numerator}/{self.denominator}"

    def __repr__(self):
        return f"Fraction({self.numerator}, {self.denominator})"

    @classmethod
    def from_float(cls, value):
        if isinstance(value, int):
            return cls(value, 1)
        str_value = str(value)
        if 'e' in str_value:
            base, exp = str_value.split('e')
            decimal_places = len(base.split('.')[1]) if '.' in base else 0
            exp = int(exp)
            numerator = int(float(base) * (10 ** (decimal_places - exp)))
            denominator = 10 ** decimal_places
        else:
            decimal_places = len(str_value.split('.')[1])
            numerator = int(value * (10 ** decimal_places))
            denominator = 10 ** decimal_places
        return cls(numerator, denominator)
# Создание дробей
f1 = Fraction(3, 4)
f2 = Fraction(5, 6)

# Сложение дробей
f3 = f1 + f2
print(f3)  # Ожидаемый вывод: 19/12

# Вычитание дробей
f4 = f1 - f2
print(f4)  # Ожидаемый вывод: -1/12

# Умножение дробей
f5 = f1 * f2
print(f5)  # Ожидаемый вывод: 5/8

# Деление дробей
f6 = f1 / f2
print(f6)  # Ожидаемый вывод: 9/10

# Проверка равенства
print(f1 == Fraction(6, 8))   # Ожидаемый вывод: True

# Сравнение дробей
print(f1 > f2)    # Ожидаемый вывод: False
print(f1 < f2)    # Ожидаемый вывод: True

# Преобразование к float
print(float(f1))  # Ожидаемый вывод: 0.75

# Обратная дробь
f7 = f1.reciprocal()
print(f7)         # Ожидаемый вывод: 4/3

# Создание дроби из float
f8 = Fraction.from_float(0.75)
print(f8)         # Ожидаемый вывод: 3/4

# Проверка обработки исключений
try:
    f_invalid = Fraction(5, 0)
except ValueError as e:
    print(e)  # Ожидаемый вывод: Знаменатель не может быть нулем

# Строковое и официальное представление
print(str(f1))    # Ожидаемый вывод: 3/4
print(repr(f1))   # Ожидаемый вывод: Fraction(3, 4)
````
